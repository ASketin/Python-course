### Теорминимум

```python
input_value = int(input("Введите число: "))

if input_value % 2 == 0:
    print('Число четное')
else:
    print("Число нечетное")
```

Допустимые операторы сравнения:
<ul>
    <li>== равно</li>
    <li>!= не равно</li>
    <li>< меньше</li>
    <li>> больше</li>
    <li>>= больше или равно</li>
    <li><= меньше или равно</li>
</ul>

```python
eyes = int(input("Введите число глаз: "))
legs = int(input("Введите число ног: "))

if eyes >= 8:
    if legs >= 8:
        print('Это паук, АААААА!!!')
    else:
        print('Морской гребешок')
else:
    if legs == 6:
        print('Это жук')
    else:
        print("Наконец, кот")
```

```python
eyes = int(input("Введите число глаз: "))
legs = int(input("Введите число ног: "))

if eyes == 40 and legs == 40:
    print('Лучше не смотреть на это чудовище!')
```

Допустимые логические операторы:
<ul>
    <li>Оператор "И" - and</li>
    <li>Оператор "ИЛИ" - or</li>
    <li>Оператор отрицания - not</li>
</ul>

```python
your_buddy = input("Кто позвонил? ")

if your_buddy == 'Святослав':
    print("Идем на ярмарку")
elif your_buddy == 'Иван':
    print('Идем в пивнуху')
else:
    print('Идем в торговый центр')
```
